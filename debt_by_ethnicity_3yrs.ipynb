```python
import matplotlib.pyplot as plt

from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

import numpy as np
```

```python
#getting yearly percent increase
#returns percentage increase
def percent_diff(f,i):
    return (((f - i) / i) * 100)



#white
y = [11.84,13.56,16.05,15.55,19.77,20.31,23.00]

#black
z = [7.22,11.52,11.36,15.32,16.47,21.27,30.00]

yy = []
zz = []

x = 0

while(x < len(y) - 1):
    yy.append(percent_diff(y[x+1] , y[x]))
    zz.append(percent_diff(z[x+1] , z[x]))



    x = x + 1


width = 0.4

xx = [2004,2007,2010,2013,2016,2019]

bar1 = np.arange(len(xx))
bar2 = [i+width for i in bar1]



plt.bar(bar1,yy , width, label = "White Borrowers")
plt.bar(bar2,zz , width,  label=  "Black Borrowers")
plt.xticks(bar1 + width/2,xx)
plt.title("Median education installment loans by race/ethnicity, percentage increase from three years prior")
plt.ylabel("%")
plt.xlabel("Year")
plt.legend()




```
