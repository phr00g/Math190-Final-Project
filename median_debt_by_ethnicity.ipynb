```python
import matplotlib.pyplot as plt

from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
```

```python
x = [2001,2004,2007,2010,2013,2016,2019]

#white people
y = [11.84,13.56,16.05,15.55,19.77,20.31,23.00]

#black people
z = [7.22,11.52,11.36,15.32,16.47,21.27,30.00]

#hispanic people
g = [7.22,6.78,12.34,12.96,14.28,19.14,17.60]

fig, ax = plt.subplots()


ax.plot(x,y,'-bo' , label="White Borrowers")
ax.plot(x,z,'-go' , label = "Black Borrowers")
ax.plot(x,g, '-ro' , label="Hispanic Borrowers")

ax.xaxis.set_major_locator(MultipleLocator(3))
plt.title("Median education installment loans by race/ethnicity")
plt.ylabel("US$ Thousands")
plt.legend()
```
