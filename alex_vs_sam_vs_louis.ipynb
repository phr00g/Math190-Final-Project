```python
import matplotlib.pyplot as plt

from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

import numpy as np
```

```python
# Alex -- private loan -- simple interest -- %9 -- $50,000 -- 72 months

principle = 50000
flat_rate = 0.08
daily_rate = flat_rate/365

abs_total = 0 
monthly_payment = 300

#interest owed is $0 on day 0 of loan
interest_balance = 0 

#total owed after each month
alex_monthly_totals = []

#literal array of months
months = []

i = 0

for i in range(1,73): #
    months.append(i)
    
    print("After month {}".format(i))


    daily_interest = daily_rate * principle
    monthly_interest = daily_interest * 30


    #past interest balance plus interest for this month
    interest_balance += monthly_interest

    print("Coming into this month, Principle Balance: {}".format(principle))
    print("This month's interest is: {}".format(monthly_interest))
    print("Your interest balance is now: {} ".format(interest_balance))
    
    
    print("You paid : {}".format(monthly_payment))

    #payment eclipses interest balance
    if monthly_payment >= interest_balance:
        

        total = (principle + interest_balance) - monthly_payment

        principle = total

        taken_from_principle = monthly_payment - interest_balance
        
        interest_balance = 0

        
        print("You paid off this month's interest, and now your interest balance is 0 and you shaved ${} off of your principle".format(taken_from_principle))
    #payment is less than interest balance
    else:
        interest_balance = interest_balance - monthly_payment
        print("You paid ${} and now your interest balance is ${}".format(monthly_payment,interest_balance))
        
    abs_total = principle + interest_balance

    print("Your total still owed is ${}".format(abs_total))

    alex_monthly_totals.append(abs_total)

    print("-------------------------------")

```
